* id, identificador  de la solicitud, serial

```
%%sql
create table Solicitud(
    id serial,
    idCliente int,
    idPedido int
);
```

## PREGUNTA 3
Crear una **VISTA** (view) llamada "Valor" conteniendo el valor total del pedido  del carrito (1 tupla) y desplegarla, la vista contiene los siguientes atributos:
* idCliente, el número identificador del cliente
* Valor, consistente en el valor total del pedido

```
%%sql
create view Valor as
select c.idCliente, sum(p.precio*c.cantidad) as Valor
from carrito c inner join plato p on c.idPlato = p.id;
```

```
%%sql drop trigger InsertaSolicitud on pedido;
```

```
%%sql

create or replace function insertaSolicitud() returns trigger
language PLPGSQL
AS
$$
BEGIN
  insert into solicitud(idCliente, idPedido)
  values(new.cliente, new.id);
  return new;

END;
$$

```

```
%%sql
create trigger InsertaSolicitud
after insert on pedido
for each row
execute procedure insertaSolicitud();
```

## Pregunta 5
Crear un **STORED PROCEDURE** llamado "ListaPedidos" que genere una lista de todos los pedidos de un cliente ingresado como parámetro y desplegarla ordenado por fecha (no hora)



```
%%sql

create or replace function ListaPedidos(id_cliente int)
returns table(
    id int,
    cliente int,
    plato varchar(100),
    fecha date,
    hora time
)
language plpgsql
as $$
BEGIN

  return query
  select *
  from pedido
  where pedido.cliente = id_cliente
  order by fecha;

END; $$
```

```
%%sql

select * from ListaPedidos(1);
```
